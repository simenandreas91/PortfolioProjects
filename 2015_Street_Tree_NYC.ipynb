# Data cleaning

Let us apply something to the nan values. But just the nan values of the stump and dead trees. The other nan values might be caused by something else.

```python
mask = ((trees_subset['status'] =='Stump') | (trees_subset['status'] == 'Dead'))
```

```python
trees_subset.loc[mask] = trees_subset.loc[mask].fillna('Not Applicable')
```

```python
trees_subset[trees_subset['status'] == 'Stump']
```

```python
trees_subset.isnull().sum()
```

We are still having some null values. Let us take a look

```python
trees_subset[trees_subset['health'].isnull()]
```

```python
trees_subset[trees_subset['problems'].isnull()].head()
```

```python
trees_subset['problems'].value_counts()
```

## From our exploration, the NaN values of 'problems' are most likely a value 'None'.

The NaN 'Health' problem is only one instance. It is hard to know what the cause might be, so we will set it to the most common condition which is 'Good'.

We are doing the same thing for NaN of 'sidewalk'.

The NaN values for the tree sorts are we going to set to 'No Observation'.

```python
trees_subset['problems'].fillna('None', inplace=True)
trees_subset['health'].fillna('Good', inplace=True)
trees_subset['spc_latin'].fillna('No Observation', inplace=True)
trees_subset['sidewalk'].fillna('NoDamage', inplace=True)
```

```python
trees_subset['sidewalk'].value_counts()
```

## Let us move all the trees with tree_dbh of 0 to the 25% mark.

## And do the same for the biggest ones.

### This is just an example of how we can do it. I believe the stats are correct

```python
stats = trees_subset_alive.groupby('spc_latin')['tree_dbh'].describe().reset_index()[['spc_latin', '25%', '75%']]
stats
```

```python
trees_subset_alive = trees_subset_alive.merge(stats, on='spc_latin', how='left')
trees_subset_alive
```

```python
mask = trees_subset_alive['tree_dbh'] < trees_subset_alive['25%']
trees_subset_alive.loc[mask, 'tree_dbh'] = trees_subset_alive['25%']

mask = trees_subset_alive['tree_dbh'] > trees_subset_alive['75%']
trees_subset_alive.loc[mask, 'tree_dbh'] = trees_subset_alive['75%']
```
